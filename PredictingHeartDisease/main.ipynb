```python
chestpaintype = heart.iloc[:,2].value_counts()
chestpaintype.plot(kind='bar',title='Chest Pain Type', xlabel='', ylabel='count', rot=0)
```

```python
fastingbs = heart.iloc[:,5].value_counts()
fastingbs.plot(kind='bar',title='Fasting BS', xlabel='', ylabel='count', rot=0)
```

```python
sex = heart.iloc[:,1].value_counts()
sex.plot(kind='bar',title='Sex', xlabel='', ylabel='count', rot=0)
```

```python
restingecg = heart.iloc[:,6].value_counts()
restingecg.plot(kind='bar',title='Resting ECG', xlabel='', ylabel='count', rot=0)
```

```python
exerciseangina = heart.iloc[:,8].value_counts()
exerciseangina.plot(kind='bar',title='Exercise Angina', xlabel='', ylabel='count', rot=0)
```

```python
stslope = heart.iloc[:,10].value_counts()
stslope.plot(kind='bar',title='ST Slope', xlabel='', ylabel='count', rot=0)
```

```python
heartdisease = heart.iloc[:,11].value_counts()
heartdisease.plot(kind='bar',title='Heart Disease', xlabel='', ylabel='count', rot=0)
```
